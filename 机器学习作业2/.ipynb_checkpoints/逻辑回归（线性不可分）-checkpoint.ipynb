```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#数据可视化
data = pd.read_csv('ex2data2.txt',names=['Test 1','Test 2','Accepted'])
fig,ax = plt.subplots()
ax.scatter(data[data['Accepted']==0]['Test 1'],data[data['Accepted']==0]['Test 2'],c='red',marker='x',label='y=0')
ax.scatter(data[data['Accepted']==1]['Test 1'],data[data['Accepted']==1]['Test 2'],c='blue',marker='o',label='y=1')
ax.legend()
ax.set(xlabel='emax1',ylabel='exam2')
plt.show()
```

```python
#特征映射（把x1,x2变成高次）
def feature_mapping(x1,x2,power):  #power为几次方
    data = {}
    for i in np.arange(power+1):
        for j in np.arange(i+1):
            data['F{}{}'.format(i-j,j)] = np.power(x1,i-j) * np.power(x2,j) #np.power()求幂，将其存到字典中次
    return pd.DataFrame(data)  #pd.DataFrame构造数据框（将字典变为表格）
```

```python
def get_Xy(data1,data2):
    y = data1.iloc[:,-1]  #X全要
    X = np.array(data2)
    y = np.array(y)
    y = y.reshape((len(y),1))
    return X,y
```

```python
#g(x)函数
def sigmoid_function(z):
    return 1/(1+np.exp(-z))
```

```python
#代价函数
def cost_function(X,y,theta,lamda):
    y1 = sigmoid_function(X@theta)
    first = y*np.log(y1)
    second = (1-y)*np.log(1-y1)
    reg = np.sum(np.power(theta[1:],2))*(lamda/(2*len(X)))  #theta0不取
    return -np.sum(first+second)/len(X)+reg
```

```python
#梯度下降函数
def gradient_descent_function(X,y,theta,alpha,times,lamda):
    costs = []
    for i in range(times):
        reg1 = theta[1:]*(lamda/len(X))
        reg1 = np.insert(reg1,0,values=0,axis=0)  #theta0不正则化，将theta等于0。np.insert(在哪插，插入位置，插入值，插行/列)
        y1 = sigmoid_function(X@theta)
        theta = theta - (alpha/len(y))*(X.T@(y1-y))-reg1
        cost = cost_function(X,y,theta,lamda)
        costs.append(cost)
    return theta,costs
```

```python
#计算准确率
def predict(X,theta):
    results = sigmoid_function(X@theta)  #返回一个矩阵
    ones=[]
    for result in results:
        if result > 0.5:
            i = 1
        else:
            i = 0
        ones.append(i)
    return ones
```

```python
x1 = data['Test 1']
x2 = data['Test 2']
data2 = feature_mapping(x1,x2,6)
X,y = get_Xy(data,data2)
alpha = 0.001
times = 200000
lamda = 0.001
theta = np.zeros((28,1))
theta1,costs1 = gradient_descent_function(X,y,theta,alpha,times,lamda)
```

```python
#绘制决策边界（等高线） 
x = np.linspace(-1.2,1.2,200)
xx,yy = np.meshgrid(x,x)  #xx,yy为一个200*200的矩阵,xx为以x为行向量的矩阵，yy为以x为列向量的矩阵
z = feature_mapping(xx.ravel(),yy.ravel(),6).values  #ravel矩阵向量化，多维变一维。values使其变成矩阵
zz = z@theta1
zz = zz.reshape(xx.shape)
fig,ax = plt.subplots()
ax.scatter(data[data['Accepted']==0]['Test 1'],data[data['Accepted']==0]['Test 2'],c='red',marker='x',label='y=0')
ax.scatter(data[data['Accepted']==1]['Test 1'],data[data['Accepted']==1]['Test 2'],c='blue',marker='o',label='y=1')
ax.legend()
plt.contour(xx,yy,zz,0)  #画等高线
ax.set(xlabel='emax1',ylabel='exam2')
plt.show()
y_ = np.array(predict(X,theta1))
y_ = y_.reshape(len(y),1)
acc = np.mean(y_==y)
print(acc)
```
