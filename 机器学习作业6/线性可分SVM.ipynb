## 吴恩达机器学习习题六：支持向量机（线性可分）<br><br>
- 观察C取值对决策边界的影响

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
from sklearn.svm import SVC
```

```python
#读取数据&数据可视化
data = sio.loadmat('./data/ex6data1.mat')
X = data['X']
y = data['y']
def plot_data(X,y):
    plt.scatter(X[:,0],X[:,1],c=y.flatten(),cmap='jet')  #通过y为0或1分开颜色
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.show()
plot_data(X,y)
```

```python
svc1 = SVC(C=1,kernel='linear')  #kernel为核函数(此函数为线性核函数)
svc1.fit(X,y.flatten())
```

```python
svc1.predict(X)
```

```python
svc1.score(X,y)
```

```python
def plot_boundary(model):
    
    X1,X2 = np.meshgrid(np.linspace(-0.5,4.5,500),  #均为(500,500)
                       np.linspace(1.3,5,500))
    z = np.c_[X1.flatten(),X2.flatten()]  #(25000,2)
    z = model.predict(z)  #(25000,),通过model.predict预测为1或0，输出一个列向量
    zz = z.reshape(X1.shape)  #(500,500)
    plt.contour(X1,X2,zz)  #等高线

plot_boundary(svc1)
plot_data(X,y)
```

```python
svc1 = SVC(C=100,kernel='linear')
svc1.fit(X,y.flatten())
plot_boundary(svc1)
plot_data(X,y)
```

- C不是很大的时候，可以忽略一些异常点的影响（如第一个图）
