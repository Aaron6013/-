# 吴恩达机器学习习题一：单变量线性回归<br>
- 习题：假设你是一家餐厅的CEO，正在考虑开一家分店。通过不同城市的人口和利润数据来决定在哪一个城市。

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#数据可视化
data=pd.read_csv('ex1data1.txt',names=['population','profit'])  #读取数据，pd.read_csv（文件路径，names=[列名]）
data.plot.scatter('population','profit',s=30)  #画图，data.plot.scatter（横坐标，纵坐标，c=颜色，s=点的大小）
plt.show()  #将图打印出来
```

```python
#将数据变成矩形
data.insert(0,'ones',1)  #在表前加一列1（原因看课），insert（添加列位置，列名，添加列元素）
X = data.iloc[:,:2]  #取表前两列，iloc[所取行位置，所取列位置]
y = data.iloc[:,-1]
X = np.array(X)  #将表转换为矩阵
y = np.array(y)
y = y.reshape(97,1)  #将y变成97*1的列向量
```

```python
#代价函数
def cost_function(X,y,theta):
    inner = np.power(X@theta-y,2)
    return np.sum(inner)/(2*len(y))
theta = np.zeros((2,1))  #给出初始西塔，默认为0
```

```python
#梯度下降函数
def gradient_descent_function(X,y,theta,alpha,times):
    costs = []
    for i in range(times):  #迭代times次
        theta = theta-alpha*(X.T@(X@theta-y))/len(y)
        cost = cost_function(X,y,theta)
        costs.append(cost)
    return costs,theta
alpha=0.02
times=2000
costs,theta=gradient_descent_function(X,y,theta,alpha,times)
print(theta)
```

```python
#将代价函数曲线画出
plt.plot(range(times),costs,linewidth=4)
plt.xlabel('time')
plt.ylabel('cost')
plt.title('Cost of Times')
plt.show()
```

```python
#将拟合函数和数据集表示出来
plt.scatter(X[:,-1],y,s=20,label='training data')
x = np.linspace(y.min(),y.max(),100)
y = theta[0,0]+theta[1,0]*x
plt.plot(x,y,c='red',linewidth=4,label='predict')
plt.legend()  #此函数可将两个图画到一块
plt.xlabel('population')
plt.ylabel('profit')
plt.axis([0,25,0,25])
plt.show()
```
