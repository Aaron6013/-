```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
#读取数据&数据可视化
data = sio.loadmat('./data/ex7data1.mat')
X = data['X']  #(50,2)
def plot_data(X):
    
    plt.scatter(X[:,0],X[:,1])
    plt.show()
plot_data(X)
```

```python
#均值归一化
X1 = X - np.mean(X,axis=0)  #(50,2),np.mean(X,axis=0)按列求均值 返回一个(1,2)的矩阵，X的每一行减去它
plot_data(X1)
```

```python
#计算协方差矩阵
sigma = (X1.T @ X1)/len(X1)  #(2,2)
```

```python
#计算特征向量
U,S,V = np.linalg.svd(sigma)
print(U.shape)
```

```python
#压缩得出新特征向量
U_reduce = U[:,0]  #(2,1)
U_reduce = U_reduce.reshape((2,1))
z = X1 @ U_reduce  #(50,1)
```

```python
#图像还原
X_appox = z @ U_reduce.T  #(50,2)
plt.scatter(X1[:,0],X1[:,1])
plt.scatter(X_appox[:,0],X_appox[:,1],c='red')
```
