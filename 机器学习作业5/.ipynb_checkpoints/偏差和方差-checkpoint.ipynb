## 了解算法性能中的方差和偏差概念<br><br>
- 利用水库水位变化预测大坝出水量

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
from scipy.optimize import minimize
```

```python
#读取数据
data = sio.loadmat('ex5data1.mat')
#训练集数据
X_train = data['X']  #(12,1)
y_train = data['y']  #(12,1)
#交叉验证集数据
X_val = data['Xval']  #(21,1)
y_val = data['yval']  #(21,1)
#测试集数据
X_test = data['Xtest']  #(21,1)
y_test = data['ytest']  #(21,1)
```

```python
#数据可视化
def data_plot(X,y):
    
    fig,ax = plt.subplots()
    ax.scatter(X,y)
    ax.set(xlabel = 'change in water level(x)',
          ylabel = 'water flowing out og the dam(y)')
    plt.show()
data_plot(X_train,y_train)
```

```python
#代价函数（含正则化）
def cost_function(theta,X,y,lamda):
    
    cost = np.sum(np.power((X@theta-y.flatten()),2))
    reg = theta[1:] @ theta[1:] *lamda
    return (cost + reg )/(2*len(X))
```

```python
#梯度向量
def reg_gradient(theta,X,y,lamda):
    
    first = ((X.T@(X@theta-y.flatten())))/len(X)
    reg = theta[1:]*(lamda/len(X))
    reg = np.insert(reg,0,values=0,axis=0)
    return first + reg
```

```python
#优化函数
def train_model(X,y,lamda):
     
        init_theta = np.zeros((X.shape[1]))
        result = minimize(fun=cost_function,
                         x0=init_theta,
                         args=(X,y,lamda),
                         method='TNC',
                         jac=reg_gradient)
        return result.x
```

```python
lamda = 1
X_train = np.insert(X_train,0,values=1,axis=1)  #(12,2)
X_val = np.insert(X_val,0,values=1,axis=1)  #(21,2)
X_test = np.insert(X_test,0,values=1,axis=1)  #(21,2)
theta = np.ones((X_train.shape[1]))  #(2,)
final_theta = train_model(X_train,y_train,lamda)
#画出训练集拟合函数
plt.scatter(X_train[:,1:],y_train)
x = np.linspace(-60,40,100)
xx = x.reshape(100,1)
xx = np.insert(xx,0,1,axis=1)
y1 = xx@final_theta
plt.plot(x,y1,c='red')
plt.show()
```

- 此时次数低，欠拟合

## 训练样本从1开始递增进行训练，比较训练集和交叉验证集上的损失函数的变化情况

```python
#画出学习曲线
def plot_learning(X_train,y_train,X_val,y_val,lamda):
    
    train_costs = []
    val_costs = []
    for i in range(1,len(X_train)+1):
        X_train_1 = X_train[:i,:]
        y_train_1 = y_train[:i,:]
        final_theta = train_model(X_train_1,y_train_1,lamda)
        train_cost = cost_function(final_theta,X_train_1,y_train_1,lamda)
        val_cost = cost_function(final_theta,X_val,y_val,lamda)
        train_costs.append(train_cost)
        val_costs.append(val_cost)
    plt.plot(range(1,len(X_train)+1),train_costs,c='red',label='train_costs')
    plt.plot(range(1,len(X_train)+1),val_costs,c='green',label='val_costs')
    plt.legend()
    plt.xlabel('number of training examples')
    plt.ylabel('error')
    plt.show()
plot_learning(X_train,y_train,X_val,y_val,lamda=0)      
```

- 由上图可以看出，训练集误差和交叉验证集误差刚开始相差很大，随着m增大，训练集误差小幅增大，交叉训练集误差减小，两个误差逐渐相等。结合之前的图，可得出此图为欠拟合随m增大情况。

 ## 任务：构造多项式特征，进行多项式回归

```python
#特征映射
def ploy_feature(X,power):
    
    for i in range(2,power+1):
        X = np.insert(X,X.shape[1],values=np.power(X[:,1],i),axis=1)
    return X
```

```python
#特征缩放
def get_mean_std(X):
    
    means = np.mean(X,axis=0)
    stds = np.std(X,axis=0)
    X[:,1:] = ( X[:,1:]  - means[1:]) / stds[1:]
    return X 
```

```python
X_train_feature = ploy_feature(X_train,6)
X_val_feature = ploy_feature(X_val,6)
X_test_feature = ploy_feature(X_val,6)
X_train_norm = get_mean_std(X_train_feature)
X_val_norm = get_mean_std(X_val_feature)
X_test_norm = get_mean_std(X_test_feature)
```

```python
final_theta = train_model(X_train_norm,y_train,lamda=0)
```

```python
def plot_poly_fit(X,y,final_theta):
    
    x = np.linspace(-60,60,100)
    xx = x.reshape(100,1)
    xx = np.insert(xx,0,1,axis=1)
    xx = ploy_feature(xx,6)
    xx = get_mean_std(xx)
    
    plt.plot(x, xx@final_theta,'r--')
    plt.scatter(X[:,1:],y)
    plt.show()
plot_poly_fit(X_train,y_train,final_theta)   
```

- 由此图可看出，次数增高，会过拟合

```python
plot_learning(X_train_norm,y_train,X_val_norm,y_val,lamda=0)
```

```python
plot_learning(X_train_norm,y_train,X_val_norm,y_val,lamda=1)
```

- 此时我们次数增大，过拟合，且lamda适中，此图为过拟合随m增大，两个误差都随m增大而减小且趋于一个较小的值。

```python
plot_learning(X_train_norm,y_train,X_val_norm,y_val,lamda=100)
```

```python
def plot_lamda(X1,X2,y1,y2):
    lamdas = [0,0.001,0.003,0.01,0.03,0.1,0.3,1,3,10]
    train_costs = []
    val_costs = []
    for lamda in lamdas:
        final_theta1 = train_model(X1,y1,lamda)
        train_cost = cost_function(final_theta1,X1,y1,lamda)
        val_cost = cost_function(final_theta1,X2,y2,lamda)
        train_costs.append(train_cost)
        val_costs.append(val_cost)
    plt.plot(lamdas,train_costs,c='red',label='train costs')
    plt.plot(lamdas,val_costs,c='green',label='val costs')
    plt.xlabel('lamda')
    plt.ylabel('costs')
    plt.legend()
    plt.show()
plot_lamda(X_train_norm,X_val_norm,y_train,y_val)
```

- 由上图可以看出，随着lamda增大（过拟合到欠拟合），训练集误差在增大，交叉验证集误差先减小后增大。
